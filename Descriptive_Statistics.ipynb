**Q7: Variance for Decision-Making (Medium)**

Two delivery companies track delivery delays (in minutes).

- **Company A:** variance = **6**
- **Company B:** variance = **15**

 Which company is more consistent, and why?

 **Ans-** Given:

- **Company A:** variance = **6**
- **Company B:** variance = **15**

 **Company A is more consistent**

 **Why ?**

 Variance measures how spread out the data is from the average.

- A **smaller variance** means the delivery delays are **more consistent** and closer to the average time.

- A **larger variance** means delivery delays vary more and are **less predictable**.

**Company A is more consistent** because it has a **lower variance**, meaning its delivery delays are less spread out and more predictable than Company B’s.

**Q9: Combining Measures (Hard)**

A family records their monthly electricity usage (in kWh): [400, 420, 390, 450, 410].
Find the mean and standard deviation. What do these values together tell you about the family’s energy use pattern?

**Ans-** Given Electricity Usage (in kWh):

**[400, 420, 390, 450, 410]**

- **Step-1:** **Find the mean**.

  **FORMULA:** Mean = Sum of observations / No. of observations
    
    = 400+420+390+450+410 / 5  = **414**

    **Mean = 414 kWh**

- **Step-2:** **Find the Standard Deviation**.

  | Values | Deviation from Mean | Squared Deviation |
  |--------|---------------------|-------------------|
  |  400   |  400 - 414 = -14    |    196            |
  |  420   |  420 - 414 = 6      |    36             |
  |  390   |  390 - 414 = -24    |    576            |
  |  450   |  450 - 414 = 36     |    1296           |
  |  410   |  410 - 414 = -4     |    16             |

  Total Squared Deviations = **2120**
   
- **Variance:** = Total Squared Deviations / Sum of observations

    = 2120 / 5 = **424**

  **Variance = 424**

- **Standard Deviation:**  = Square root(Variance)

    = Square root(424) ≈ 20.59

  **Standard Deviation ≈ 20.59 kWh**

  **What do these values tell us together?**  

  - The **mean electricity usage** is **414 kWh**, and **the standard deviation** is about **20.59 kWh**. Together, these values show that the family usually uses around **414 kWh per month**, with **moderate variation** in electricity consumption.
